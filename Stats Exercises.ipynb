```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

#Create a discrete random variable (drv)
drv = stats.poisson(mu = 5) #poissonian distribution
x_axis = np.arange(0, 20)

fig, axis_drv = plt.subplots(3, 1, sharex = True)
axis_drv[0].step(x_axis, drv.pmf(x_axis))
axis_drv[1].step(x_axis, drv.cdf(x_axis))
axis_drv[2].hist(drv.rvs(size = 1000))
plt.show()

#Create a continious random variable (crv)
crv_y = stats.norm() #normal distribution
crv_x = np.arange(-10, 10, 0.1)
figs_crv, axis_crv = plt.subplots(3, 1, sharex = True) #create 3 plots
axis_crv[0].plot(crv_x, crv_y.pdf(crv_x)) #plot its probability mass function (PMF)
axis_crv[1].plot(crv_x, crv_y.cdf(crv_x)) #cummulative distribution function (CDF)
axis_crv[2].hist(crv_y.rvs(size = 1000), bins = crv_x) #histogram of 1000 random realisations
plt.show()

#test
test1, p1 = stats.ttest_ind(crv_y.rvs(size = 1000), crv_y.rvs(size = 1000))
test2, p2 = stats.ttest_ind(crv_y.rvs(size = 1000), np.random.normal(loc = 0.15, size = 1000))

print(test1, p1)
print(test2, p2)
```
